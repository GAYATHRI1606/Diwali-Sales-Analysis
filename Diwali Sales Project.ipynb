```python
Group2=df.groupby('Occupation')[['Amount','Orders']].sum().reset_index()
Group2=Group2.sort_values('Amount',ascending=False)
```

```python
fig,ax1=plt.subplots(figsize=(22,14),facecolor='#F7F0CA')
sns.barplot(data=Group2,x='Occupation',y='Amount',color='#F28729',ax=ax1,label='Total Sales',legend=False)
plt.title('Occupation-wise Sales and Order Comparison',fontsize=35,fontweight='bold',color='#702254')
ax1.set_xlabel('Occupation',fontsize=25)
ax1.set_ylabel('Total Amount',fontsize=25)
ax1.set_facecolor('#F7F0CA')
ax2=ax1.twinx()
sns.lineplot(data=Group2,x='Occupation',y='Orders',color='#A31E21',marker='o',markeredgecolor='#A31E21',ax=ax2,linewidth=3,label='Total Orders',legend=False)
ax2.set_ylabel('Total Orders',fontsize=25)
fig.legend(bbox_to_anchor=(1,1),bbox_transform=ax1.transAxes,fontsize=23)
plt.savefig('Occupation.png')
plt.show()
```

```python
Group5 = df.groupby('Product_Category')['Amount'].sum().reset_index()
Group5 = Group5.sort_values('Amount', ascending=False)

top1=Group5.nlargest(5,'Amount')
top1 = Group5.head(5)

plt.figure(figsize=(12,6), facecolor='#F7F0CA')

labels = top1['Product_Category'] + "\n₹" + top1['Amount'].round().astype(int).astype(str)

squarify.plot(
    sizes=top1['Amount'],
    label=labels,
    pad=True,
    color=['#E65100','#F57C00','#FF9E0F','#F4B558','#DAD17A'],
    text_kwargs={'fontsize':12, 'color':'white'}
)

plt.title('Top 5 Best-Selling Product Categories', fontsize=18, fontweight='bold', color='#702254')
plt.axis(False)
plt.savefig('Products.png')
plt.show()
```

***Targeted Marketing Campaign Strategy:***

*The analysis shows that:*

- **Gender:** Female

- **Most Purchased Categories:** Food, Clothing, Footwear, Electronics  

*State-wise Target*

- **Uttar Pradesh :**
  Females aged 26–35, mostly working in Aviation, prefer Food the most.
  Promote festive food hampers + Buy 1 Get 1 offers through Instagram & WhatsApp ads.

- **Maharashtra :**
  Females aged 26–35, mostly working in Healthcare, prefer Food the most.
  Focus on grocery discount sales, festival bundles, and cashback offers on digital payments.

- **Karnataka :**
  Females aged 26–35, mostly working in IT Sector, prefer Footwear & Shoes the most.
  Use online fashion sale campaigns, influencer try-on videos, and limited time festival offers.

*Conclusion*

Targeting Females (26–35) with state-wise product messaging and occupation-aware promotions will significantly increase festive sales performance and customer engagement.
